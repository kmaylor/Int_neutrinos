```python
#Preambule
import numpy as np
import matplotlib.pyplot as plt
import pyfits, os
from matplotlib import rcParams
%matplotlib inline

#Directory Structure
os.chdir('/home/kmaylor/int_neutrinos/cosmoslik/cosmoslik_plugins/models/camb/camb/spectra')

# Generic properties from matplotlib
rcParams['font.family'] = 'serif'
rcParams['text.usetex'] = True
```

```python
#File roots to plot
rootplot = ['test','test2']#['Base_TT_TE_EE_lowl','test','test2']
labelplot = [r'$f_{\rm int} = 0$',r'$f_{\rm int} = 1/3$']#[r'$\Lambda{\rm CDM}$',r'$f_{\rm int} = 0$',r'$f_{\rm int} = 1/3$']
```

```python
#Function to set figure size
def figscale(left_margin=1.06, right_margin=0.18, 
	bottom_margin=0.83, top_margin=0.28, 
	axhsize=6.76, axvsize=4.89):
    """This function finds the canvas size in inches, and axis params
    in figure coords for an axis with the given offsets and size 
    (in inches)"""

    fighsize_in = left_margin + axhsize + right_margin
    figvsize_in = bottom_margin + axvsize + top_margin

    left = left_margin/fighsize_in
    bottom = bottom_margin/figvsize_in
    width = axhsize/fighsize_in
    height = axvsize/figvsize_in

    return ((fighsize_in, figvsize_in), [left, bottom, width, height])
```

```python
#Load Lensed Cls
scalCls = []
for index,root in enumerate(rootplot):
    temp = np.genfromtxt(root +'_lensedCls.dat',dtype=[('l','i8'),('TT','f8'),('EE','f8'),('BB','f8'),('TE','f8')])
    scalCls.append(temp)
```

```python
#Extract Spectra
#Initialize arrays
l=[]
ClsTT=[]
ClsTTl2=[]
ClsEE=[]
ClsTE=[]
ClsBB=[]

#Loop over spectra
for index in range(len(rootplot)):
    l.append(scalCls[index]['l'])
    ClsTT.append(scalCls[index]['TT'])
    ClsTTl2.append(scalCls[index]['l']**2*scalCls[index]['TT'])
    ClsEE.append(scalCls[index]['EE'])
    ClsTE.append(scalCls[index]['TE'])
    ClsBB.append(scalCls[index]['BB'])
```

```python
#Load data
root = '/Users/Francis-yan/Work/dark_sector/'
WMAP9TT =  np.genfromtxt(root+'CMB_data/wmap_binned_tt_spectrum_9yr_v5.txt',dtype=[('l','f8'),('lmin','i8'),('lmax','i8'),('TT','f8'),('dTT','f8'),('dTT1','f8'),('dTT2','f8')])
WMAP9EE =  np.genfromtxt(root+'CMB_data/wmap_binned_ee_spectrum_9yr_v5.txt',dtype=[('l','f8'),('lmin','i8'),('lmax','i8'),('EE','f8'),('dEE','f8'),('dEE1','f8'),('dEE2','f8')])
SPT =  np.genfromtxt(root+'CMB_data/cmb_bandpowers_spt.dat',dtype=[('l','f8'),('TT','f8'),('dTT','f8')])
ACT =  np.genfromtxt(root+'CMB_data/ACT_cl_3yr_v1.dat',dtype=[('l','f8'),('TT','f8'),('dTT','f8')])
filePlanck = pyfits.open(root+'CMB_data/COM_PowerSpect_CMB_R2.02.fits')
```

```python
#Extract Data
#Planck TT High-ell
Planck2 = filePlanck[7].data
lPlanck2 = np.int32(Planck2['ELL'])
TTPlanck2 =  Planck2['D_ELL']
dTTPlanck2 = Planck2['ERR']
TTl2Planck2 =  np.power(lPlanck2,2)*Planck2['D_ELL']
dTTl2Planck2 = np.power(lPlanck2,2)*Planck2['ERR']

#Planck TT Low-ell
Planck_low_l = filePlanck[1].data
lPlanck_low = np.int32(Planck_low_l['ELL'])
TTPlanck_low = Planck_low_l['D_ELL']
dUTTPlanck_low = Planck_low_l['ERRUP']
dDTTPlanck_low = Planck_low_l['ERRDOWN']

#Planck EE Low-ell
Planck_low_l_EE = filePlanck[3].data
lPlanck_low_EE = np.int32(Planck_low_l_EE['ELL'])
EEPlanck_low = Planck_low_l_EE['D_ELL']
dUEEPlanck_low = Planck_low_l_EE['ERRUP']
dDEEPlanck_low = Planck_low_l_EE['ERRDOWN']

#Planck EE High-ell
Planck2EE = filePlanck[11].data
lPlanck2EE = np.int32(Planck2EE['ELL'])
EEPlanck2 =  Planck2EE['D_ELL']
dEEPlanck2 = Planck2EE['ERR']

#Planck TE Low-ell
Planck_low_l_TE = filePlanck[2].data
lPlanck_low_TE = np.int32(Planck_low_l_TE['ELL'])
TEPlanck_low = Planck_low_l_TE['D_ELL']
dUTEPlanck_low = Planck_low_l_TE['ERRUP']
dDTEPlanck_low = Planck_low_l_TE['ERRDOWN']

#Planck TE High-ell
Planck2TE = filePlanck[9].data
lPlanck2TE = np.int32(Planck2TE['ELL'])
TEPlanck2 =  Planck2TE['D_ELL']
dTEPlanck2 = Planck2TE['ERR']

#WMAP TT
WlTT = WMAP9TT['l']
WTT = WMAP9TT['TT']
WdTT = WMAP9TT['dTT']

#WMAP EE
WlEE = WMAP9EE['l']
WlEEmin = WMAP9EE['lmin']
WlEEmax = WMAP9EE['lmax']
WEE = WMAP9EE['EE']
WdEE = WMAP9EE['dEE']

#SPT
Sl = SPT['l'][24:]
ST = Sl**2*SPT['TT'][24:]
SdT= Sl**2*SPT['dTT'][24:]

#ACT
ACTl = ACT['l'][13:]
ACTTT = ACTl**2*ACT['TT'][13:]
ACTdTT = ACTl**2*ACT['dTT'][13:]
```

```python
#Define the properties of the figure
(aleft_in, aright_in) = (1.5, 0.4)
(abottom_in, atop_in) = (0.75, 0.2)
(xsize_in, ysize_in) = (10, 7.0)

sizes, coords = figscale(aleft_in, aright_in, abottom_in, 
            atop_in, xsize_in, ysize_in)

cusls = ['b-','r--','k-','c-','m--']
```

```python
# Initialize the figure
fig = plt.figure()
fig.set_size_inches(sizes)
ax = fig.add_axes(coords)
# Choose figure properties
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlim(0,3000)
#ax.set_ylim(-0.12,0.02)
plt.xlabel('$l$',size = 25)
plt.ylabel('$l^3(l+1)C_l^{TT}/2\pi$ $[\mu K^2]$',size = 25)
ticklabels = ax.get_xticklabels()
ticklabels.extend( ax.get_yticklabels() )
for label in ticklabels:
    label.set_color('k')
    label.set_fontsize(25)
    label.set_fontname('Times New Roman')

#Loop over spectra
for index in range(len(rootplot)):
    plt.plot(l[index],ClsTTl2[index],cusls[index],linewidth=1.4,label=labelplot[index])

#Plot the data
#ax.errorbar(lPlanck2,TTl2Planck2,dTTl2Planck2,fmt='sg', ms=4, capsize=3,elinewidth=1,label=r'$\rm Planck$')
#ax.errorbar(Sl,ST,SdT,fmt='Dy', ms=4, capsize=3,elinewidth=1,label=r'$\rm SPT$')
#ax.errorbar(ACTl,ACTTT, ACTdTT,fmt='^c', ms=4, capsize=3,elinewidth=1,label = r'$\rm ACT$')

#The legend
ax.legend(loc = 'best',fontsize=20)
```
