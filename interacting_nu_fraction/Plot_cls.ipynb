```python
#Preambule
import numpy as np
import matplotlib.pyplot as plt
import pyfits, os
from matplotlib import rcParams
%matplotlib inline

#Directory Structure
os.chdir('/home/kmaylor/interacting_nu_fraction/spectra')

# Generic properties from matplotlib
rcParams['font.family'] = 'serif'
rcParams['text.usetex'] = True
```

```python
#File roots to plot
rootplot = ['test','test2']#['Base_TT_TE_EE_lowl','test','test2']
labelplot = [r'$f_{\rm int} = 0$',r'$f_{\rm int} = 1/3$']#[r'$\Lambda{\rm CDM}$',r'$f_{\rm int} = 0$',r'$f_{\rm int} = 1/3$']
```

```python
#Function to set figure size
def figscale(left_margin=1.06, right_margin=0.18, 
	bottom_margin=0.83, top_margin=0.28, 
	axhsize=6.76, axvsize=4.89):
    """This function finds the canvas size in inches, and axis params
    in figure coords for an axis with the given offsets and size 
    (in inches)"""

    fighsize_in = left_margin + axhsize + right_margin
    figvsize_in = bottom_margin + axvsize + top_margin

    left = left_margin/fighsize_in
    bottom = bottom_margin/figvsize_in
    width = axhsize/fighsize_in
    height = axvsize/figvsize_in

    return ((fighsize_in, figvsize_in), [left, bottom, width, height])
```

```python
#Load Lensed Cls
scalCls = []
for index,root in enumerate(rootplot):
    temp = np.genfromtxt(root +'_lensedCls.dat',dtype=[('l','i8'),('TT','f8'),('EE','f8'),('BB','f8'),('TE','f8')])
    scalCls.append(temp)
```

```python
#Extract Spectra
#Initialize arrays
l=[]
ClsTT=[]
ClsTTl2=[]
ClsEE=[]
ClsTE=[]
ClsBB=[]

#Loop over spectra
for index in range(len(rootplot)):
    l.append(scalCls[index]['l'])
    ClsTT.append(scalCls[index]['TT'])
    ClsTTl2.append(scalCls[index]['l']**2*scalCls[index]['TT'])
    ClsEE.append(scalCls[index]['EE'])
    ClsTE.append(scalCls[index]['TE'])
    ClsBB.append(scalCls[index]['BB'])
```

```python
#Define the properties of the figure
(aleft_in, aright_in) = (1.5, 0.4)
(abottom_in, atop_in) = (0.75, 0.2)
(xsize_in, ysize_in) = (10, 7.0)

sizes, coords = figscale(aleft_in, aright_in, abottom_in, 
            atop_in, xsize_in, ysize_in)

cusls = ['b-','r--','k-','c-','m--']
```

```python
# Initialize the figure
fig = plt.figure()
fig.set_size_inches(sizes)
ax = fig.add_axes(coords)
# Choose figure properties
#ax.set_xscale('log')
#ax.set_yscale('log')
ax.set_xlim(0,3000)
#ax.set_ylim(-0.12,0.02)
plt.xlabel('$l$',size = 25)
plt.ylabel('$l^3(l+1)C_l^{TT}/2\pi$ $[\mu K^2]$',size = 25)
ticklabels = ax.get_xticklabels()
ticklabels.extend( ax.get_yticklabels() )
for label in ticklabels:
    label.set_color('k')
    label.set_fontsize(25)
    label.set_fontname('Times New Roman')

#Loop over spectra
for index in range(len(rootplot)):
    plt.plot(l[index],ClsTTl2[index],cusls[index],linewidth=1.4,label=labelplot[index])

#Plot the data
#ax.errorbar(lPlanck2,TTl2Planck2,dTTl2Planck2,fmt='sg', ms=4, capsize=3,elinewidth=1,label=r'$\rm Planck$')
#ax.errorbar(Sl,ST,SdT,fmt='Dy', ms=4, capsize=3,elinewidth=1,label=r'$\rm SPT$')
#ax.errorbar(ACTl,ACTTT, ACTdTT,fmt='^c', ms=4, capsize=3,elinewidth=1,label = r'$\rm ACT$')

#The legend
ax.legend(loc = 'best',fontsize=20)
```
